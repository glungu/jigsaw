```python
from IPython.display import display
def show_by_column_threshold(column_name, threshold):
    records_n = df[pd.isnull(df[column_name])]
    records_z = df[df[column_name] == 0.0]
    records_l = df[df[column_name] <= threshold]
    records_g = df[df[column_name] > threshold]
    records_zt = records_z[records_z['target'] > 0.5]
    records_nt = records_n[records_n['target'] > 0.5]
    records_lt = records_l[records_l['target'] > 0.5]
    records_gt = records_g[records_g['target'] > 0.5]
    
    print('Number of ' + column_name + ' == None:', 
          (100. * len(records_n)/len(df)), '%', 
          ', with target > 0.5: ',  
          (100. * len(records_nt)/len(records_n)) if len(records_n) > 0 else 0.0, '%')

    print('Number of ' + column_name + ' == 0.0:', 
          (100. * len(records_z)/len(df)), '%', 
          ', with target > 0.5: ',  
          (100. * len(records_zt)/len(records_z)), '%')
    
    print('Number of ' + column_name + ' <= ' + str(threshold) + ':', 
          (100. * len(records_l)/len(df)), '%', 
          ', with target > 0.5: ',  
          (100. * len(records_lt)/len(records_l)), '%')

    print('Number of ' + column_name + ' > ' + str(threshold) + ':', 
          (100. * len(records_g)/len(df)), '%', 
          ', with target > 0.5: ',  
          (100. * len(records_gt)/len(records_g)), '%')

    display(records_g[['id','target','comment_text',column_name]].head())
```

```python
show_by_column_threshold('target', 0.5)
```

```python
from datetime import datetime
# 2015-09-29 10:50:52.451277+00
def str_to_date(s):
    try:
        result = datetime.strptime(s + '00', '%Y-%m-%d %H:%M:%S.%f%z')
    except ValueError:
        result = datetime.strptime(s + '00', '%Y-%m-%d %H:%M:%S%z')
    return result

df['created_date_num'] = df['created_date'].apply(str_to_date)
df[['created_date_num']].describe() 
```
